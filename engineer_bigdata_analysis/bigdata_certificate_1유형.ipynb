#### T1-19. 시계열 데이터3

주어진 데이터에서 2022년 월별 Sales 합계 중 가장 큰 금액과
2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값으로 구하시오.

단 Events컬럼이 '1'인경우 80%의 Salse값만 반영함

(최종값은 소수점 반올림 후 정수 출력)

데이터셋 : basic2.csv

```python
df = pd.read_csv('/content/drive/MyDrive/빅분기 실기/basic2.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
```

```python
def event_sales(x):
  if x['Events'] == 1:
    x['Sales2'] = x['Sales'] * 0.8
  else:
    x['Sales2'] = x['Sales']

  return x

df = df.apply(lambda x: event_sales(x), axis=1)
```

```python
df2 = df[df['year'] == 2022]
sales22 = df2.groupby('month')['Sales2'].sum().max()
sales22
```

```python
df3 = df[df['year'] == 2023]
sales23 = df3.groupby('month')['Sales2'].sum().max()
sales23
```

```python
print(int(round(abs(sales22 - sales23), 0)))
```

```python
df['previous_PV'] = df['PV'].shift(1)
df.head()
```

```python
df['previous_PV'] = df['previous_PV'].fillna(method='bfill')
df.head()
```

#### t1-91-sigmoid

[심화 학습] basic1 데이터에서 f2 값에 시그모이드 함수(그림 내 수식)를 적용하고, f4가 ISFJ인 (변경된)f2의 합을 구하시오! (소수점 둘째자리까지 출력, 반올림)

```python
df = pd.read_csv('/content/drive/MyDrive/빅분기 실기/basic1.csv')
df.head()
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

df['f2_sigmoid'] = df['f2'].apply(sigmoid)

print(round(df[df['f4'] == 'ISFJ']['f2_sigmoid'].sum(), 2))
```

#### t1-92-relu

basic1 데이터에서 age 값에 ReLU 함수를 적용하고, city가 부산인 (변경된)age의 평균을 구하시오! (소수점 둘째자리까지 출력, 절사)

```python
df = pd.read_csv('/content/drive/MyDrive/빅분기 실기/basic1.csv')
df.head()
```

```python
def relu(x):
  return np.maximum(0, x)

df['age_relu'] = df['age'].apply(relu)

print(int(df[df['city'] == '부산']['age_relu'].mean()))
```
