```python
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'
df5 = pd.read_csv(url)
df5.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df5.head()
```

```python
df5.shape
```

```python
dic = {
    'Unknown' : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K' : 'b',
    '$60K - $80K' : 'c',
    '$80K - $120K' : 'd',
    '$120K +' : 'e'
}
                    # list(map(lambda x: dic[x], df5['Income_Category']))
df5['newIncome'] = df5.Income_Category.map(lambda x: dic[x])
ans = df5['newIncome']
ans.head()
```

```python
def changeCategory(x):
    if x == 'Unknown':
        return 'N'
    elif x == 'Less than $40K':
        return 'a'
    elif x == '$40K - $60K':
        return 'b'
    elif x == '$60K - $80K':
        return 'c'
    elif x == '$80K - $120K':
        return 'd'
    elif x == '$120K +':
        return 'e'
    
df5['newIncome'] = df5.Income_Category.apply(changeCategory)
ans = df5['newIncome']
ans.head()
```

```python
df5['AgeState'] = df5.Customer_Age.map(lambda x: x//10 *10)
ans = df5['AgeState'].value_counts().sort_index()
ans
```

```python
df5['newEduLevel'] = df5.Education_Level.map(lambda x: 1 if 'Graduate' in x else 0)
ans = df5['newEduLevel'].value_counts()
ans
```

```python
df5['newLimit'] = df5.Credit_Limit.map(lambda x: 1 if x >= 4500 else 0)
ans = df5['newLimit'].value_counts()
ans
```

```python
def check(x):
    if x.Marital_Status == 'Married' and x.Card_Category == 'Platinum':
        return 1
    else:
        return 0
    
df5['newState'] = df5.apply(check, axis=1)
ans = df5['newState'].value_counts()
ans
```

```python
def changeGender(x):
    if x == 'M':
        return 'male'
    elif x == 'F':
        return 'female'
    
df5['Gender'] = df5.Gender.apply(changeGender)
ans = df5['Gender'].value_counts()
ans
```

```python
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv'
df7 = pd.read_csv(url)
df7.info()
```

```python
df7.head()
```

```python
def change_date(x):
    import datetime
    
    hour = x.split(':')[1]
    date = x.split(':')[0]
    
    if hour == '24':
        hour = '00:00:00'
        
        finalDate = pd.to_datetime(date +" "+hour) + datetime.timedelta(days=1)
        
    else:
        hour = hour + ':00:00'
        
        finalDate = pd.to_datetime(date +" "+hour)
    return finalDate
df7['(년-월-일:시)'] = df7['(년-월-일:시)'].apply(change_date)
df7.head()
    
```

```python
df7['dayName'] = df7['(년-월-일:시)'].dt.day_name()
df7['dayName'].head()
```

```python
ans1 = df7.groupby(['dayName', 'PM10등급'],as_index=False).size()
ans2 = ans1.pivot(index='dayName', columns='PM10등급', values='size').fillna(0)
ans2.head()
```
