```python
df4 = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
df4.head()
```

```python
ans = df4.groupby('host_name').size()
ans.head()
```

```python
ans = df4.groupby('host_name').size().\
                to_frame().rename(columns={0:'counts'}).\
                    sort_values('counts', ascending=False)
ans.head()
```

```python
ans = df4.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
ans.head()
```

```python
ans = df4.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()\
    .groupby(['neighbourhood_group'], as_index=False).max()
ans
```

```python
ans = df4[['neighbourhood_group', 'price']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
ans
```

```python
ans = df4[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
ans
```

```python
ans = df4.groupby(['neighbourhood', 'neighbourhood_group']).price.mean()
ans
```
